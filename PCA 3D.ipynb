```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
#Generating random dataset:-
np.random.seed(234234)

#class1
mean1=np.array([0,0,0])
cov1=np.array([[1,0,0],[0,1,0],[0,0,1]])
class1=np.random.multivariate_normal(mean1,cov1,100)

#class2
mean2=np.array([1,1,1])
cov2=np.array([[1,0,0],[0,1,0],[0,0,1]])
class2=np.random.multivariate_normal(mean2,cov2,100)
```

```python
#PLotting 3D data:-
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D,proj3d
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')
ax.plot(class1[:,0],class1[:,1],class1[:,2],'o')
ax.plot(class2[:,0],class2[:,1],class2[:,2],'^')
plt.show()
```

```python
#Applying PCA:--
data=np.concatenate((class1,class2))
pca=PCA(n_components=2)
trans_data=pca.fit_transform(data)
trans_data
```

```python
pca.components_
```

```python
plt.scatter(trans_data[:,0],trans_data[:,1])
```

```python
#Inversion
approx_data=pca.inverse_transform(trans_data)
```
